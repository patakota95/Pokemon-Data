# Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)

#We cant load to the while 20GB file once, so we load 100MB at a time
#df = pd.read_csv('modified.csv', chunksize=5) #By using 5 it says that only 5 rows are being passed
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])
    #print("CHUNK DF")
   # print(df)
print(new_df)
```
